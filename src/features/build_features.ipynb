```python
import pandas as pd
import numpy as np
```

```python
df_products = pd.read_csv('../../data/raw/products_detail.csv')
df_products.head()
```

```python
df_products_copy = df_products.copy()
```

```python
# use label encoder to encode the categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_products_copy['p_categories'] = le.fit_transform(df_products_copy['p_categories'])
df_products_copy['p_brand'] = le.fit_transform(df_products_copy['p_brand'])
```

```python
df_products_copy.head()
```

```python
# scale the p_brand and p_categories
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_products_copy['p_brand'] = scaler.fit_transform(df_products_copy['p_brand'].values.reshape(-1,1))
df_products_copy['p_categories'] = scaler.fit_transform(df_products_copy['p_categories'].values.reshape(-1, 1))
df_products_copy
```

```python
canomical_data = df_products_copy.drop(['p_id','p_name', 'p_id_shop', 'p_shop_name'], axis=1)
```

```python
canomical_data
```

```python
canomical_data.to_csv('../../data/processed/canomical_data.csv', index=False)
```
