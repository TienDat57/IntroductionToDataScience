# <font color="green"><b>Question 1:</b></font>
### <strong>What is the question?</strong>
&#9889; Which store is the best seller and which item is the best seller in category is the best seller? And why is it the best seller?
### <strong>What are benefits of finding the answer?</strong>
&#9889; The answer of this question will help the company to know which store is the best seller and which item is the best seller in category is the best seller. And why is it the best seller? The company can use this information to improve the business. And check out strategies to improve the business and increase the sales.
### <strong>How to find the answers of questions by using data?</strong>
1. The first step is find category is the best seller. 
2. The second step is find store is the best seller.
3. The third step is find item is the best seller in category is the best seller. 
#### <strong>1. <i>The first step is find category is the best seller.</i> </strong>

```python
df_cate_best_seller = df_products['p_category'].value_counts().sort_values(ascending=False).index[0]
df_cate_best_seller = df_products[df_products['p_category'] == df_cate_best_seller]
df_cate_best_seller.head()
```

#### <strong>2. <i>The second step is find store is the best seller.</i> </strong>

```python
df_shop_best_seller = df_cate_best_seller['p_id_shop'].value_counts().sort_values(ascending=False).index[0]
df_shop_best_seller = df_cate_best_seller[df_cate_best_seller['p_id_shop'] == df_shop_best_seller]
df_shop_best_seller.head()
```

#### <strong>3. <i>The third step is find item is the best seller in category is the best seller.<i><strong>

```python
df_top5_products = df_shop_best_seller.sort_values(by='p_sold_quantity', ascending=False)
df_top5_products.head(10)
```

```python
# visualize top 5 products
plt.figure(figsize=(10, 5))
sns.barplot(x='p_name', y='p_sold_quantity', data=df_top5_products.head(5))
plt.xticks(rotation=45)
plt.show()
```

&#9889; <font color="yellow"><b>Conclusion:</b></font> The best seller category is "Nhà Cửa - Đời Sống" and the best seller store is "Tiki Trading" and the best seller item is "Hộp Hâm Nóng Life360 EL-ALS263 (Tím)" in category is the best seller. And the reason is the price of the item is low, discount is high, the number of orders is high (Buyer psychology will choose products with a large number of purchases to choose to buy), the quality of the item is good, brand of the item is famous (Life360, LocknLock,...), strategy advertising of the item is good.

# <font color="green"><b>Question 2:</b></font>
### <strong>What is the question?</strong>
&#9889; Sales of each category in `Tiki trading`. The best-selling category accounts for what percentage of that sales?
### <strong>What are benefits of finding the answer?</strong>
&#9889; The sales of each category in `Tiki trading`, which is the sales unit with the most number of products sold, will help the store 
adjust increase, or decrease the number of products to meet the needs of customers as well as help increase profits for stores as well as manufacturers.
### <strong>How to find the answers of questions by using data?</strong>
0. Create a sales column by multiplying the quantity of a product sold by its original selling price
1. Group each category in tiki trading, then calculate the total number of products sold by each category
2. Calculate sales by multiplying the quantity of a product sold by its original selling price
3. Calculate the percentage of sales of the best-selling category compared to the total sales of all categories


```python
df_products_copy = df_products.copy()
```

#### <strong>0. <i>Create a sales column by multiplying the quantity of a product sold by its original selling price</i></strong>

```python
df_products_copy['p_sales'] = df_products_copy['p_original_price'] * df_products_copy['p_sold_quantity']
df_products_copy
```

#### <strong>1. <i>Group each category in tiki trading, then calculate the total number of products sold by each category.</i> </strong>

```python
df_cate_in_store = df_products_copy[df_products_copy['p_shop_name'] == 'Tiki Trading'].groupby('p_category')['p_sales'].sum()
df_cate_in_store
```

-  Because category 'CHƯƠNG TRÌNH KHUYẾN MẠI KHÁC' not a category but it has high sales, so we need to handle it

```python
promotion = df_products_copy[df_products_copy['p_category'] == 'CHƯƠNG TRÌNH KHUYẾN MẠI KHÁC']
promotion   
```

```python
# we will move the product with id 232 to category 'Nhà cửa - Đời sống' and the product with id 887 to category 'Hộp đựng thực phẩm'
df_products_copy.loc[df_products_copy['p_id'] == 134621577, 'p_category'] = 'Nhà Cửa - Đời Sống'
df_products_copy.loc[df_products_copy['p_id'] == 3526073, 'p_category'] = 'Hộp đựng thực phẩm'
```

#### <strong>2. <i>Calculate sales by multiplying the quantity of a product sold by its original selling price.</i> </strong>

```python
# group each category in store which has p_shop_name is Tiki trading
df_cate_in_store = df_products_copy[df_products_copy['p_shop_name'] == 'Tiki Trading'].groupby('p_category')

# get the total sales of each category
df_cate_in_store_total_sales = df_cate_in_store['p_sales'].sum().sort_values(ascending=False).reindex().rename_axis('category').reset_index(name='total_sales')

df_cate_in_store_count = df_cate_in_store['p_id'].count().sort_values(ascending=False).reindex().rename_axis('category').reset_index(name='count')

# merge two dataframe
df_cate_in_store_total_sales = df_cate_in_store_count.merge(df_cate_in_store_total_sales, on='category')
df_cate_in_store_total_sales.loc['Sum'] = df_cate_in_store_total_sales.sum()
df_cate_in_store_total_sales
```

- As the result, we can see that the category with the highest number of products sold is 'Nhà Cửa - Đời Sống' with 44 products sold and has a sales over of 100,000,000,000 VND.

#### <strong>3. <i>Calculate the percentage of sales of the best-selling category compared to the total sales of all categories.</i> </strong>

```python
df_cate_in_store_total_sales['percentage(%)'] = round(df_cate_in_store_total_sales['total_sales'] / df_cate_in_store_total_sales['total_sales']['Sum'] * 100,2)
df_cate_in_store_total_sales
```

```python
df_cate_in_store_total_sales[:10].plot.bar(x='category', y='percentage(%)', rot=90, figsize=(10,7))
```

&#9889; <font color="yellow"><b>Conclusion:</b></font>
- 'Nhà Cửa - Đời Sống' is the category with the highest number of products sold and the highest sales over.
- The percentage of sales of the best-selling category('Nhà Cửa - Đời Sống') is more than 50% of the total sales of all categories. So, the store should focus on this category to increase sales. Because this is the category that customers are most interested in and  necessary items in daily life.

# <font color="green"><b>Question 4:</b></font>
### <strong>What is the question?</strong>
&#9889; Calculate the average daily sales of the brand. Bring out the brand with the highest and lowest revenue.
### <strong>What are benefits of finding the answer?</strong>
&#9889; The answer to this question will help big brands see if their current revenue is up to their expectations, then adjust prices, advertise brands, and come up with new policies.
### <strong>How to find the answers of questions by using data?</strong>
1. The first step is to calculate the total historical revenue of each product.
2. Then calculate the average total income of each brand in 1 day and find the brand with the largest and smallest revenue per day.
3. Finally, visualize for users to see the difference in revenue of each brand from high to low.
#### <strong>1. <i>The first step is to calculate the total historical revenue of each product.</i> </strong>

```python
# df_products1 = df_products.copy()
# df_products1['total_revenue'] = df_products1['p_sold_quantity'] * df_products1['p_original_price']
df_products_copy['total_history_revenue'] = df_products_copy['p_sales'] / df_products_copy['p_day_created']
df_products_copy['total_history_revenue'] = df_products_copy['total_history_revenue'].round(2)
df_products_copy = df_products_copy.drop(['p_id', 'p_id_shop', 'p_shop_name', 'p_category', 'p_day_created', 'p_sold_quantity', 'p_original_price', 'p_current_price', 'p_discount_rate'], axis=1)
df_products_copy.head(10)
```

#### <strong>2. <i>Then calculate the average total income of each brand in 1 day and find the brand with the largest and smallest revenue per day.</i> </strong>

```python
df_brand_revenue = df_products_copy.groupby('p_brand').sum().sort_values(by='total_history_revenue', ascending=False)
df_brand_revenue = df_brand_revenue.rename(columns={'total_history_revenue': 'average_sales_per_day'})
df_brand_revenue.head(10)
```

#### <strong>3. <i>Finally, visualize for users to see the difference in revenue of each brand from high to low.</i></strong>

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df_brand_revenue.index[:10], y=df_brand_revenue['average_sales_per_day'][:10])
plt.xticks(rotation=90)
plt.show()
```

&#9889; <font color="yellow"><b>Conclusion:</b></font> 
- From that we can see, <font color="orange">LocknLock</font> topped the sales table and doubled that of the 2nd brand. And the 2nd brand is twice the 3rd brand. Later on, the difference between brands is not too big. And here, we see that LocknLock has a revenue of nearly 200 million a day. But the data doesn't show how much of <font color="orange">LocknLock's capital</font> was spent.
- <font color="orange">Locknlock</font>'s capital can be spent a lot to have such a huge number in one day. And <font color="orange">LocknLock</font> is also one of the big brands and promotes a lot in all fields of communication: social networks, leaflets, ..... About social networks as we know today, young people and many age of use, so <font color="orange">LocknLock</font>'s daily sales will also be as large. Discounts are also an advantage for increasing sales, but in return, the profit of a product will be very low.

# <font color="green"><b>Question 5:</b></font>
### <strong>What is the question?</strong>
&#9889; What products do brands have? Compare sales of the same product but in different brands.
### <strong>What are benefits of finding the answer?</strong>
&#9889; Solving this question will help brands identify which products sell best and will have to compete with other brands' products.
### <strong>How to find the answers of questions by using data?</strong>
1. The first step is to calculate the total historical revenue of each product.
2. Then we need to group by brands with categories
3. The next, we will choose sample categories to compare with other brands based on its sales
4. 
#### <strong>1. <i>The first step is to calculate the total historical revenue of each product.</i> </strong>

```python
df_products2 = df_products.copy()
df_products2['total_revenue'] = df_products2['p_sold_quantity'] * df_products2['p_original_price']
df_products2 = df_products2.drop(['p_id', 'p_id_shop', 'p_shop_name', 'p_day_created', 'p_original_price', 'p_current_price', 'p_discount_rate'], axis=1)
df_products2.head(10)
```

#### <strong>2. <i>Then we need to group by brands with categories.</i> </strong>

```python
# group each brand with category
df_brand_cate = df_products2.groupby(['p_brand', 'p_category']).sum()
df_brand_cate = df_brand_cate.sort_values(by='total_revenue', ascending=False)
df_brand_cate.head(10)
```

#### <strong>3. <i>The next, we will choose sample categories to compare with other brands based on its sales.</i> </strong>

```python
# drop row which have p_category is not 'Nhà Cửa - Đời Sống'
df_brand_cate = df_brand_cate.drop(df_brand_cate[df_brand_cate.index.get_level_values('p_category') != 'Nhà Cửa - Đời Sống'].index)
df_brand_cate.head(10)
```

```python
#visualize top 10 brand have total revenue is the highest and p_sold_quantity is the highest
df_brand_cate[:10].plot.bar(y='p_sold_quantity', rot=90, figsize=(10,7))
df_brand_cate[:10].plot.bar(y='total_revenue', rot=90, figsize=(10,7))
```

&#9889; <font color="yellow"><b>Conclusion:</b></font>
- To answer for the difference in revenue, here we have presented a category, 'Nhà Cửa - Đời Sống' is one of the categories with the highest number of products. As we can see, although the revenue difference is large (8 times), the number of products sold is 25 times higher than that of <font color="orange">Sumika</font>. Prove, the price of <font color="orange">LocknLock</font> is very low.
- Partly because <font color="orange">LocknLock</font> promotes products to every social network, every front in the market. As for Sumika, it is expensive but not so profitable, assuming 10 people buy 10 <font color="orange">LocknLock</font> products, Sumika only needs to sell 1 product to make a profit like <font color="orange">LocknLock</font>.
