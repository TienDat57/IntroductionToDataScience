## 2.2.3. How to find the answers of questions by using data?
1. The first step is find category is the best seller.

#### 1. The first step is find category is the best seller.

```python
df_cate_best_seller = df_products['p_categories'].value_counts().sort_values(ascending=False).index[0]
df_cate_best_seller = df_products[df_products['p_categories'] == df_cate_best_seller]
df_cate_best_seller.head()
```

#### 2. The third step is find store is the best seller.

```python
df_shop_best_seller = df_cate_best_seller['p_id_shop'].value_counts().sort_values(ascending=False).index[0]
df_shop_best_seller = df_cate_best_seller[df_cate_best_seller['p_id_shop'] == df_shop_best_seller]
df_shop_best_seller.head()
```

#### 3. The second step is find item is the best seller in category is the best seller. 

```python
# get top 10 products in best seller shop in best seller category 
df_top5_products = df_shop_best_seller.sort_values(by='p_sold_quantity', ascending=False)
df_top5_products.head(10)
```

And the reason is the price of the item is low, discount is high, the number of orders is high (Buyer psychology will choose products with a large number of purchases to choose to buy), the quality of the item is good, brand of the item is famous (Life360, LocknLock,...), strategy advertising of the item is good.

# <font color="green"><b>Question 2:</b></font>

#### 2.2.1. What is the question?
&#9889; Sales of each category in `Tiki trading`. The best-selling category accounts for what percentage of that sales?

#### 2.2.2. What are benefits of finding the answer?
&#9889; The sales of each category in `Tiki trading`, which is the sales unit with the most number of products sold, will help the store

#### 2.2.3. How to find the answers of questions by using data?
0. Create a sales column by multiplying the quantity of a product sold by its original selling price
1. Group each category in tiki trading, then calculate the total number of products sold by each category
2. Calculate sales by multiplying the quantity of a product sold by its original selling price
3. Calculate the percentage of sales of the best-selling category compared to the total sales of all categories

```python
# make a copy of the dataframe
df_products_copy = df_products.copy()
```

#### 0. Create a sales column by multiplying the quantity of a product sold by its original selling price

```python
# create a new column to store the sales by multiplying the price and the sold quantity
df_products_copy['p_sales'] = df_products_copy['p_original_price'] * df_products_copy['p_sold_quantity']
df_products_copy
```

#### 1. Group each category in tiki trading, then calculate the total number of products sold by each category

```python
df_cate_in_store = df_products_copy[df_products_copy['p_shop_name'] == 'Tiki Trading'].groupby('p_categories')['p_sales'].sum()
df_cate_in_store
```

-  Because category 'CHƯƠNG TRÌNH KHUYẾN MẠI KHÁC' not a category but it has high sales, so we need to handle it

```python
# take the product which have category is 'CHƯƠNG TRÌNH KHUYẾN MẠI KHÁC'
promotion = df_products_copy[df_products_copy['p_categories'] == 'CHƯƠNG TRÌNH KHUYẾN MẠI KHÁC']
promotion   
```

```python
# we will move the product with id 232 to category 'Nhà cửa - Đời sống' and the product with id 887 to category 'Hộp đựng thực phẩm'
df_products_copy.loc[df_products_copy['p_id'] == 134621577, 'p_categories'] = 'Nhà Cửa - Đời Sống'
df_products_copy.loc[df_products_copy['p_id'] == 3526073, 'p_categories'] = 'Hộp đựng thực phẩm'
```

```python
# group each category in store which has p_shop_name is Tiki trading
df_cate_in_store = df_products_copy[df_products_copy['p_shop_name'] == 'Tiki Trading'].groupby('p_categories')

# get the total sales of each category
df_cate_in_store_total_sales = df_cate_in_store['p_sales'].sum().sort_values(ascending=False).reindex().rename_axis('category').reset_index(name='total_sales')
df_cate_in_store_total_sales

df_cate_in_store_count = df_cate_in_store['p_id'].count().sort_values(ascending=False).reindex().rename_axis('category').reset_index(name='count')
df_cate_in_store_count

# merge two dataframe
df_cate_in_store_total_sales = df_cate_in_store_count.merge(df_cate_in_store_total_sales, on='category')
df_cate_in_store_total_sales.loc['Sum'] = df_cate_in_store_total_sales.sum()
df_cate_in_store_total_sales
```

- As the result, we can see that the category with the highest number of products sold is 'Nhà Cửa - Đời Sống' with 44 products sold and has a sales over of 100,000,000,000 VND.

```python
# the percentage of each category in store
df_cate_in_store_total_sales['percentage(%)'] = round(df_cate_in_store_total_sales['total_sales'] / df_cate_in_store_total_sales['total_sales']['Sum'] * 100,2)
df_cate_in_store_total_sales
```

```python
# show bar chart with top 10 categories in store
df_cate_in_store_total_sales[:10].plot.bar(x='category', y='percentage(%)', rot=90, figsize=(10,7))
```

- Comments:
    - 'Nhà Cửa - Đời Sống' is the category with the highest number of products sold and the highest sales over.
    - The percentage of sales of the best-selling category('Nhà Cửa - Đời Sống') is more than 50% of the total sales of all categories. So, the store should focus on this category to increase sales. Because this is the category that customers are most interested in and  necessary items in daily life.

#### 1. We reuse the data from the previous question to find the total sales of each category `df_cate_best_seller`

#### 2. We find the total sales of top 5 item have number of orders that discount is 0 and top 5 item have number of orders that discount is not 0 in category is the best seller

```python
# get product which have discount rate is 0
product_not_discount = df_products_copy[df_products_copy['p_discount_rate'] == 0]
# get top 5 product have sold quantity is the highest
product_not_discount = product_not_discount.sort_values(by='p_sold_quantity', ascending=False)[:5]
product_not_discount
```

```python
# get product which have discount rate is not 0
product_discount = df_products_copy[df_products_copy['p_discount_rate'] != 0]
# get top 5 product have sold quantity is the highest
product_discount = product_discount.sort_values(by='p_sold_quantity', ascending=False)[:5]
product_discount
```

#### 3. We find shop is the best seller in category is the best seller

```python
# find shop of product_discount and product_not_discount 
product_discount_shop = product_discount['p_shop_name'].unique()
product_not_discount_shop = product_not_discount['p_shop_name'].unique()
print('Shop of product_discount: ', product_discount_shop)
print('Shop of product_not_discount: ', product_not_discount_shop)
```

#### 4. Visualize the result by using bar chart -> Compare the total sales of top 5 item have number of orders that discount is 0 and top 5 item have number of orders that discount is not 0 in category is the best seller

```python
# Visualize the result by using bar chart -> Compare the total sales of top 5 item have number of orders that discount is 0 and top 5 item have number of orders that discount is not 0 in category is the best seller
plt.figure(figsize=(10,7)) 
plt.title('Top 5 product have sold quantity is the highest in category is the best seller')
plt.bar(product_discount['p_name'], product_discount['p_sales'], color='blue', label='Discount')
plt.bar(product_not_discount['p_name'], product_not_discount['p_sales'], color='red', label='Not discount')
plt.xticks(rotation=90)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

&#9889; <font color="yellow"><b>Conclusion:</b></font> Looking at the data visualize we see that the total sales of top 5 item have number of orders that discount is 0 is higher than the total sales of top 5 item have number of orders that discount is not 0. Discounted products have more sales than non-discounted products, a sizable difference in sales. This is because the price of discounted products is lower than the price of non-discounted products. So, customers will choose to buy discounted products. And the reason why non-discounted products sell more expensive than discounted products of the same item is because the quality of the product is better than the discounted product. And the store should focus on selling non-discounted products to increase sales. The reasons for this disparity include:
- The quality of the product is better than the discounted product.
